#  Plotting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Cereal Data - Sheet1.csv')
df.set_index('name', inplace=True)

%matplotlib notebook
%matplotlib inline

print(df.head())
```

```python
# plt.style.available
```

```python
df.plot('sugars', 'calories', kind = 'scatter');
```

```python
fit = np.polyfit(df.sugars,df.calories, 1)
fit_fn = np.poly1d(fit) 


# calculate the y values based on the co-efficients from the model
r_x, r_y = zip(*((i, i*fit[0] + fit[1]) for i in df.sugars))
 
# Put in to a data frame, to keep is all nice
lm_original_plot = pd.DataFrame({
'sugars' : r_x,
'calories' : r_y
})
fig, axes = plt.subplots()

df.plot(x='sugars', y='calories', kind = 'scatter', ax = axes, title='Original Values')
lm_original_plot.plot(x='sugars', y='calories', kind = 'line', ax=axes, color = 'red')

xsum = 0
k = 0
for i in df.calories:
    k += 1
    xsum += i - fit_fn[k]
print(xsum/k)
plt.show()
```
